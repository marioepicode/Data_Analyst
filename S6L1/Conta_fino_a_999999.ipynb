```python
# -*- coding: iso-8859-15 -*-

# Scrivere un algoritmo che, dato un numero,
# ne mostri la sua rappresentazione a lettere in italiano
# Esempio:
#   input: 1234 -> output: milleduecentotrentaquattro
#
# Come funziona?
# Per i primi venti numeri, non c' altra strada che quella
# di prevedere una traduzione semplice attraverso una tabella:
# 0 -> zero, 1 -> uno, 2 -> due, ..., 19 -> diciannove

def translate_to_20(n):
    if n > 19:
        return "Out of range"

    NUMBERS = ["","uno", "due", "tre", "quattro", "cinque", "sei", "sette",
               "otto", "nove", "dieci", "undici", "dodici", "tredici",
               "quattordici", "quindici", "sedici", "diciassette",
               "diciotto", "diciannove"]
    return NUMBERS[n]

# dal 20, fino al 100 (escluso), ho la possibilit di prevedere
# una "traduzione" della decina e demandare la "traduzione"
# dell'unit alla funzione che traduce fino a 20
# 25 -> decina = 2, unit = 5




def translate_to_100(n):
    if n < 20:
        return translate_to_20(n)
    if n > 99:
        return "Out of range"
    
    DECADES = ["venti", "trenta", "quaranta", "cinquanta", "sessanta",
               "settanta", "ottanta", "novanta"]
    decade = n // 10
    unit = n % 10
    
    if unit == 1 or unit == 8:
        return DECADES[decade - 2][:-1] + translate_to_20(unit)
    else:
        return DECADES[decade - 2] + translate_to_20(unit)

def translate_to_100_2(n):
    if n < 20:
        return translate_to_20(n)
    if n > 99:
        return "Out of range"
    
    DECADES = ["venti", "trenta", "quaranta", "cinquanta", "sessanta",
               "settanta", "ttanta", "novanta"]
    decade = n // 10
    unit = n % 10
    
    if unit == 1 or unit == 8:
        return DECADES[decade - 2][:-1] + translate_to_20(unit)
    else:
        return DECADES[decade - 2] + translate_to_20(unit)





def translate_to_1000(n):
    if n < 100:
        return translate_to_100(n)
    if n > 999:
        return "Out of range"
    
    hundreds = n // 100
    remainder = n % 100
    
    if remainder == 0:
        if hundreds == 1:
            return "cento"
        else:
            return translate_to_20(hundreds) + "cento"
    else:
        if hundreds == 1:
            return "cento" + translate_to_100_2(remainder)
        else:
            return translate_to_20(hundreds) + "cento" + translate_to_100_2(remainder)
        
    

def translate_to_100000(n):
    if n < 1000:
        return translate_to_1000(n)
    if n > 999999:
        return "Out of range"

    thousands = n // 1000
    remainder = n % 1000

    if remainder == 0:
        if thousands == 1:
            return "mille"
        else:
            return translate_to_1000(thousands) + "mila"
    else:
        if thousands == 1:
            return "mille" + translate_to_1000(remainder)
        else:
            return translate_to_1000(thousands) + "mila" + translate_to_1000(remainder)

    
    
    
def translate_number(n):
    return translate_to_100000(n)

for x in range(1, 10000):
    print(translate_number(x))




```
